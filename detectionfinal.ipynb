# Originial Dataset 320x320

```python
model = YOLO('yolov8n.pt')
model.train(data='/content/drive/MyDrive/PSP_aug_yolo/data.yaml',
            imgsz=320,
            epochs=100,
            batch=32,
            device=0,
            project='supernova_sr_exp',
            name='yolov8n_320')
```

```python
!yolo val model=supernova_sr_exp/yolov8n_320/weights/best.pt \
         data=/content/drive/MyDrive/PSP_aug_yolo/data.yaml imgsz=320
```

```python
!yolo predict model=supernova_sr_exp/yolov8n_320/weights/best.pt \
             source=/content/drive/MyDrive/PSP_aug_yolo/images/test \
             imgsz=320 save_txt save_conf
```
